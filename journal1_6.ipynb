```python
import numpy as np
import math
#from astropy.table import QTable
```

```python
def ValuesSine(x):
    return math.sin(x)

def ValuesCosine(x):
    return math.cos(x)
```

```python
for i in range(1000):
    x = 2 * math.pi * i / 999  
    s = ValuesSine(x)
    c = ValuesCosine(x)
    print(f"{s:.3f}\t\t{c:.3f}\t\t{x:.3f}")
    
```

```python
for i in range(10):
    x = 2 * math.pi * i / 999  
    s = ValuesSine(x)
    c = ValuesCosine(x)
    print(f"x\t\ts\t\tc")
    print(f"{x:.3f}\t\t{s:.3f}\t\t{c:.3f}")
    
```
