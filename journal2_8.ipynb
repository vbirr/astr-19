```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# array from [0,1] in 100 steps
x = np.linspace(0, 1, 100)

# functionns that return sin(x) and sin(x)
y = np.sin(x)
z = np.cos(x)
```

```python
f, arr = plt.subplots(1,2)

# first panel
arr[0].plot(x,y)
arr[0].set_xlabel('x')
arr[0].set_ylabel('sin(x)')

# second panel
arr[1].plot(x,z)
arr[1].set_xlabel('x')
arr[1].set_ylabel('cos(x)')

# adding some space between plots to increase legibility
f.subplots_adjust(wspace = 0.4)

f.savefig('session_8.pdf')
```
